```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import os

# 1. Configuration
save_path = "model_legend.png"
colors = {
    'Fusion': '#8E44AD',
    'Two-Stream': '#2980B9',
    'CNN-RNN': '#D35400',
    'I3D': '#27AE60'
}

# 2. Setup Figure (Invisible canvas)
# We make it small because we only care about the legend box
fig, ax = plt.subplots(figsize=(4, 3))
ax.axis('off')  # Hide the graph part

# 3. Create Legend Elements
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Fusion',
           markerfacecolor='#8E44AD', markersize=18),
    Line2D([0], [0], marker='o', color='w', label='Two-Stream',
           markerfacecolor='#2980B9', markersize=18),
    Line2D([0], [0], marker='o', color='w', label='CNN-RNN',
           markerfacecolor='#D35400', markersize=18),
    Line2D([0], [0], marker='o', color='w', label='I3D',
           markerfacecolor='#27AE60', markersize=18)
]

# 4. Draw the Legend
# fontname='Times New Roman' ensures it matches your manual edits
legend = ax.legend(handles=legend_elements, loc='center',
                   fontsize=14, frameon=True, edgecolor='gray')

# Optional: Set the font family for the text specifically if needed
plt.setp(legend.get_texts(), fontname='Times New Roman')

# 5. Save
plt.savefig(save_path, dpi=300, bbox_inches='tight', transparent=True)
print(f"Legend saved to: {os.path.abspath(save_path)}")

# Automatic Download (for Colab)
try:
    from google.colab import files
    files.download(save_path)
except ImportError:
    pass
```
